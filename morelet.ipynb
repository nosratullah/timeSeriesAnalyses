```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

```python
freq = 10
time = np.linspace(-1,1,101)
tetha = np.pi/2.

std = .2
```

```python
sin = np.sin(2 * np.pi * freq * time + tetha)
gauss = np.exp(-(time/std)**2/2)
wavelet = sin * gauss
```

```python
#plt.plot(time,sin, alpha = 0.5);
#plt.plot(time, gauss, alpha = 0.5);
plt.plot(time, wavelet)
```
